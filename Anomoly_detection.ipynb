Importing dataset

```python
from pycaret.datasets import get_data
anomaly = get_data('anomaly')
anomaly.shape
```

```python
anomaly.describe()
```

Importing anomaly detection module

```python
from pycaret.anomaly import *
exp_ano = setup(anomaly)
```

Creating a model

```python
knn = create_model('knn')
```

Plotting the model

```python
plot_model(knn)
```

```python
svm = create_model('svm')
plot_model(svm)
```

```python
knn_predictions = predict_model(knn, data = anomaly)
knn_predictions
```

```python
svm_predictions = predict_model(svm, data = anomaly)
svm_predictions
```

```python
knn_results = assign_model(knn)
knn_results
```

```python
svm_results = assign_model(svm)
svm_results
```

Performing Classification

```python
from pycaret.classification import *
clf1 = setup(data = knn_results, target = 'Anomaly')
clf2 = setup(data = svm_results, target = 'Anomaly')
```

Comparing all the models

```python
#returns the best model
best = compare_models(sort = 'Recall',)
```
